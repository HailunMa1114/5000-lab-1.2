# Data Science and Analytics at Georgetown

Georgetown University M.S. in Data Science & Analytics (DSAN) program provides driven individuals with the in-depth technical skills and career development support needed to thrive in the high-growth data science field. The curriculum allows students to focus on areas like deep learning, big data, cloud computing, advanced Natural Language Processing (NLP), Large Language Models (LLMs), AI and more. Access to hands-on research opportunities and expert faculty actively working in the data science field gives students practice in real-world skill application, making students a valuable addition to any industry. Whether students enroll part-time or full-time, join us to study Data Science & Analytics at Georgetown to transform students' passion and knowledge into expertise, navigating the data-driven world with confidence.
